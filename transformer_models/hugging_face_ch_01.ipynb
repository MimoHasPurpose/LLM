```python
from transformers import pipeline
```

```python
classifier=pipeline("sentiment-analysis")
classifier("i been waiting for a stressfree life for some time!!")
```

```python
classifier([
    "hi",
    "whatever"
])
```

```python
classifier=pipeline('zero-shot-classification')
classifier("this is a course abut death", candidate_labels=["education","death","anarchy"],)
```

```python
generator=pipeline("text-generation")
```

```python
generator("I am")
```

```python
generator=pipeline("text-generation",model="gpt2")
```

```python
generator(
    "in this mindset",
    max_length=30,
    num_return_sequences=2,
)
```

```python
unmasker=pipeline('fill-mask')
unmasker("this course is <mask> shitty",top_k=2)
```

```python
# fill-mask pipeline will predict mising words in a sentence
```

```python
ner=pipeline("ner",grouped_entities=True)
ner("my name is a good boi boi boi")
```

```python
# NER: identifies persons orgs, locations
```
